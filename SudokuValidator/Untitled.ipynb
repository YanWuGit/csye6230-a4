```python
# Read the content of the sudokuInput file to see if it's in the correct format
file_path = 'sudokuInput'

# Function to read the sudoku puzzle from the input file
def read_sudoku_input(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            # Assuming the file has a single line with numbers separated by spaces
            numbers = list(map(int, content.split()))
            if len(numbers) != 81:
                return "Incorrect number of digits in the file. Expected 81, got {}.".format(len(numbers))
            return numbers
    except Exception as e:
        return "An error occurred: {}".format(e)

# Read the file and check the format
sudoku_numbers = read_sudoku_input(file_path)
sudoku_numbers
```
